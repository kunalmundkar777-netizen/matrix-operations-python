```python
"""
Project Title : Matrix Operations Using Python and NumPy
Author        : Kunal Jaydas Mhunkar
"""

import numpy as np


# --------------------------------
# Function to print matrix neatly
# --------------------------------
def print_matrix(M):
    for row in M:
        print(row)


# --------------------------------
# Function to input matrix
# --------------------------------
def input_matrix(name):
    rows = int(input(f"Enter number of rows for Matrix {name}: "))
    cols = int(input(f"Enter number of columns for Matrix {name}: "))

    matrix = []

    print(f"Enter elements row-wise for Matrix {name}:")

    for i in range(rows):
        row = list(map(float, input().split()))
        matrix.append(row)

    return np.array(matrix)


# --------------------------------
# Main Program
# --------------------------------
while True:

    print("\n===== MATRIX OPERATIONS =====")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Transpose")
    print("5. Determinant")
    print("6. Inverse")
    print("7. Rank")
    print("0. Exit")

    choice = input("\nEnter your choice: ")

    if choice == "0":
        print("Program exited successfully.")
        break

    # Input first matrix
    A = input_matrix("A")

    # Input second matrix if required
    if choice in ["1", "2", "3"]:
        B = input_matrix("B")

    print("\n----- RESULT -----")

    # Addition
    if choice == "1":
        if A.shape == B.shape:
            print_matrix(A + B)
        else:
            print("Matrices must have same dimensions.")

    # Subtraction
    elif choice == "2":
        if A.shape == B.shape:
            print_matrix(A - B)
        else:
            print("Matrices must have same dimensions.")

    # Multiplication
    elif choice == "3":
        if A.shape[1] == B.shape[0]:
            print_matrix(np.dot(A, B))
        else:
            print("Columns of A must equal rows of B.")

    # Transpose
    elif choice == "4":
        print_matrix(A.T)

    # Determinant
    elif choice == "5":
        if A.shape[0] == A.shape[1]:
            print("Determinant =", np.linalg.det(A))
        else:
            print("Determinant only exists for square matrices.")

    # Inverse
    elif choice == "6":
        if A.shape[0] == A.shape[1]:
            try:
                print_matrix(np.linalg.inv(A))
            except np.linalg.LinAlgError:
                print("Inverse does not exist.")
        else:
            print("Inverse only exists for square matrices.")

    # Rank
    elif choice == "7":
        print("Rank =", np.linalg.matrix_rank(A))

    else:
        print("Invalid choice.")
```
